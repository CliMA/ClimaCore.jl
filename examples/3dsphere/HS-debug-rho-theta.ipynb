```python
import numpy as np
import netCDF4 as nc
import matplotlib.pyplot as plt
```

```python
data = nc.Dataset("hsdata/debug-rho_theta_rll.nc")
```

```python
lon = data.variables['lon'][:].data
lat = data.variables['lat'][:].data
t = data.variables['time'][:].data
z = data.variables['z'][:].data
rho = data.variables['rho'][:].data
theta = data.variables['theta'][:].data
u = data.variables['u'][:].data
v = data.variables['v'][:].data
```

# last steps

```python
_fig, _ax = plt.subplots(3,5, figsize=(35, 18))

for it in np.arange(5):
    uct = _ax[0, it].contourf(lat, z, u[-(it+1), :,:,:].mean(axis=2), vmin = -30, vmax = 30, cmap='RdBu_r')
    _fig.colorbar(uct, ax=_ax[0, it])
    vct = _ax[1, it].contourf(lat, z, v[-(it+1), :,:,:].mean(axis=2), vmin = -5, vmax = 5, cmap='RdBu_r')
    _fig.colorbar(vct, ax=_ax[1, it])
    
    thetact = _ax[2, it].contour(lat, z, theta[-(it+1), :,:,:].mean(axis=2), levels=np.arange(270,450,10))
    _ax[2, it].clabel(thetact)

    _ax[0, it].set_title("u at last "+str(it)+" step")
    _ax[1, it].set_title("v at last "+str(it)+" step")
    _ax[2, it].set_title("theta at last "+str(it)+" step")

plt.show()
```

```python
_fig, _ax = plt.subplots(3,5, figsize=(35, 18))

for it in np.arange(5):
    uct = _ax[0, it].contourf(lat, z, u[-(it+1+5), :,:,:].mean(axis=2), vmin = -30, vmax = 30, cmap='RdBu_r')
    _fig.colorbar(uct, ax=_ax[0, it])
    vct = _ax[1, it].contourf(lat, z, v[-(it+1+5), :,:,:].mean(axis=2), vmin = -5, vmax = 5, cmap='RdBu_r')
    _fig.colorbar(vct, ax=_ax[1, it])
    
    thetact = _ax[2, it].contour(lat, z, theta[-(it+1+5), :,:,:].mean(axis=2), levels=np.arange(270,450,10))
    _ax[2, it].clabel(thetact)

    _ax[0, it].set_title("u at last "+str(it+5)+" step")
    _ax[1, it].set_title("v at last "+str(it+5)+" step")
    _ax[2, it].set_title("theta at last "+str(it+5)+" step")

plt.show()
```
